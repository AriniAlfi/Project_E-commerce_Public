```python
monthly_orders_df = sellers_order_items_df.resample(rule = 'M', on = 'shipping_limit_date').agg({
    "order_id" : "nunique",
    "total" : "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_id" : "order_count",
    "total" : "revenue"
}, inplace = True)
monthly_orders_df.head(20)
```

```python
monthly_orders_df = pd.DataFrame(monthly_orders_df)
monthly_orders_df["shipping_limit_date"] = pd.to_datetime(monthly_orders_df["shipping_limit_date"])
monthly_orders_df=monthly_orders_df[monthly_orders_df["shipping_limit_date"].dt.year == 2016]
plt.figure(figsize=(10,5))
plt.plot(
    monthly_orders_df["shipping_limit_date"],
    monthly_orders_df["revenue"],
    marker = 'o',
    linewidth = 1,
    color = "#72BCD4"
)
plt.title("Total Revenue Order Item per Month (2016)",loc = "center", fontsize = 20)
plt.xlabel("shipping_limit_date", fontsize=12)
plt.ylabel("revenue", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```
